```python
#Calculator
from tkinter import *

c=Tk()
exp=""

#configure
c.title("Calculator")
c.geometry('600x400')

def exp_f(num):
    global exp
    exp=exp+str(num)
    s.set(exp)
    
def clear():
    global exp
    exp=""
    s.set(exp)
    
def result():
    try:
        global exp
        ans=str(eval(exp))
        s.set(ans)
        exp=" "
    except:
        s.set("Error!")
        exp=" "
    
s=StringVar()    
#application name
name=Label(c,text="SYSTEM CALCULATOR",font=("Times New Roman",20,"bold","underline"),fg="black")
name.place(x=50,y=10)


#input
inp=Entry(c,width=70,textvariable=s)
inp.place(x=5,y=100)

#buttons
b1=Button(c,text="1",command=lambda:exp_f(1))
b1.place(x=10,y=150)

b2=Button(c,text="2",command=lambda:exp_f(2))
b2.place(x=40,y=150)

b3=Button(c,text="3",command=lambda:exp_f(3))
b3.place(x=70,y=150)

b4=Button(c,text="4",command=lambda:exp_f(4))
b4.place(x=100,y=150)

b5=Button(c,text="5",command=lambda:exp_f(5))
b5.place(x=130,y=150)

b6=Button(c,text="6",command=lambda:exp_f(6))
b6.place(x=10,y=200)

b7=Button(c,text="7",command=lambda:exp_f(7))
b7.place(x=40,y=200)

b8=Button(c,text="8",command=lambda:exp_f(8))
b8.place(x=70,y=200)

b9=Button(c,text="9",command=lambda:exp_f(9))
b9.place(x=100,y=200)

b0=Button(c,text="0",command=lambda:exp_f(0))
b0.place(x=130,y=200)

add=Button(c,text="+",command=lambda:exp_f('+'))
add.place(x=10,y=250)

sub=Button(c,text="-",command=lambda:exp_f('-'))
sub.place(x=40,y=250)

mul=Button(c,text="*",command=lambda:exp_f('*'))
mul.place(x=70,y=250)

div=Button(c,text="/",command=lambda:exp_f('/'))
div.place(x=100,y=250)

equal=Button(c,text="=",command=result)
equal.place(x=130,y=250)

dot=Button(c,text=".",command=lambda:exp_f('.'))
dot.place(x=10,y=300)

clear=Button(c,text="C",command=clear)
clear.place(x=40,y=300)

c.mainloop()
```
