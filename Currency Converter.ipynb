```python
#Currency Converter
from tkinter import *

c=Tk()
op={"American Dollar":76.22,"Euro":86.10,"British Pounds":101.05,"Chinese Yaun":11.96,"Japanese Yen":0.667,"AED":20.74,
    "Kuwaiti Dinar":251.18,"SGD":55.80,"Russian Ruble":1.034,"Australian Dollar":54.22,"Canadian Dollar":58.97,
    "Singapore Dollar":55.55,"Swiss Franc":82.29,"Venezuelan Bolivar":0.000165,"Indonesian Rupiah":0.00529,
    "Colombian Peso":0.0191,"South Korean Won":0.0639,"Nepalese Rupee":0.624}

#configure
c.title("Currency Converter")
c.geometry('800x600')

def conv():
    rupees=inp.get()
    currency=v.get()
    #result.insert(INSERT,currency)
    
    try:
        ans=float(rupees)*float(op[currency])
        result.delete(1.0,END)
        result.insert(INSERT," ",INSERT,currency,INSERT," : ",INSERT,ans)
    except:
        result.insert(INSERT,"Please select the currency")
        
#application name
name=Label(c,text="Currency Converter",font=("Lucida Calligraphy",20,"bold","underline"),fg="red")
name.place(x=250,y=10)

#label
l=Label(c,text="Enter the value in Indian Rupees",font=("Copperplate gothic bold",15,"bold"),fg="black")
l.place(x=40,y=100)

#user input
inp=Entry(c,width=35)
inp.place(x=350,y=105)

#label
l=Label(c,text="Choose the currency",font=("Copperplate gothic bold",15,"bold"),fg="black")
l.place(x=40,y=150)

#currency list
v=StringVar(c)
v.set("---Select---")
option=OptionMenu(c,v,*op)
option.place(x=350,y=150,width=220)

#result
result=Text(c,height=3,width=50)
result.place(x=350,y=200)

#button
b=Button(c,text="Convert",fg="black",font=("Copperplate gothic bold",14),bg="grey",command=conv)
b.place(x=250,y=300)

c.mainloop()
```
